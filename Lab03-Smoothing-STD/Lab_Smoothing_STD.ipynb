```python
'''
(1) Your own median filter algorithm
'''
def median_(width,height,img):
    ## copy and np.ones --> using of np.ones image is darker
    filtered_image = np.copy(img)

    edgex = width//2
    edgey = height//2
    
    img_h = img.shape[0]   # height 
    img_w = img.shape[1]   # width 
    arr = np.ones((width,height))

    for x in range(edgey,img_h - edgey):
        for y in range (edgex,img_w - edgex):
            arr = img[x-edgex:x+edgex+1, y-edgey:y+edgey+1]
            # print(arr)
            filtered_image[x][y] = np.median(arr)
            # print(filtered_image[x][y])
    show_images([img,filtered_image],["original","filtered"])

for img in ["pout.tif","tire.tif"]:  
    img = io.imread(img)
    noisy = random_noise(img,'s&p',amount=0.05)        
    median_(3,3,noisy)
```

```python
'''
(2) skimage median filter
'''
for img in ["pout.tif","tire.tif"]:
    img = io.imread(img)
    noisy = random_noise(img,'s&p',amount=0.05)  
    filtered = median(noisy)
    show_images([img,noisy,filtered],["original","noisy","median-filtered image"])
```

```python
'''
(3) Gaussion Filters
'''
from skimage.filters import gaussian

def Gaussian_filter(image):
    img = io.imread(image)
    noisy = random_noise(img,'s&p',amount=0.05)  
    gauss_img1=gaussian(noisy, 8)
    gauss_img2=gaussian(noisy, 3)
    gauss_img3=gaussian(noisy, 0.2)
    gauss_img4=gaussian(noisy, 1.87879)
    show_images([img, noisy, gauss_img1,gauss_img2,gauss_img3,gauss_img4],["original","noisy","gaussian filter with sig = 8","gaussian filter with sig = 3","gaussian filter with sig = 0.2","gaussian filter with sig = acceptable"])

for img in ["pout.tif","tire.tif"]:
    Gaussian_filter(img)
```
